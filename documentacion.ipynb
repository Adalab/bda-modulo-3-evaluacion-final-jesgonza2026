# Descripción de los Datos

El conjunto de datos está compuesto por **dos ficheros en formato csv** que, en conjunto, permiten analizar el comportamiento del cliente (vuelos, puntos, actividad) con su perfil (demografía, ingresos, valor, fidelización) dentro de un programa de lealtad de una aerolínea.

- Customer Flight Analysis: Se centra en la **actividad de vuelos y uso de puntos**

- Customer Loyalty History: Describe el **perfil del cliente y detalles sobre su membresía en el programa de lealtad**

Ambos datasets se pueden relacionar mediante la columna **Loyalty Number**.

---

## ✈️ Customer Flight Analysis.csv


- **Loyalty Number**  
  Identificador único de cada cliente. Permite relacionar este dataset con el de historial.

- **Year**
    Año en el cual se registraron las actividades de vuelo para el cliente.

- **Month**
  Representa el mes del año (de 1 a 12) en el cual ocurrieron las actividades de vuelo.

- **Flights Booked**  
  Número de vuelos reservados por el cliente en un mes específico.

- **Flights with Companions**  
  Número de vuelos reservados en los que el cliente viajó acompañado.

- **Total Flights**  
    Número total de vuelos realizados por el cliente (incluye histórico).

- **Distance**  
  Distancia total volada en el mes (en millas o kilómetros).

- **Points Accumulated**  
  Puntos acumulados por el cliente durante el mes, generalmente en función de la distancia u otros factores.

- **Points Redeemed**  
  Puntos que el cliente ha utilizado (canjeado) en ese mes.

- **Dollar Cost Points Redeemed**  
  Coste en dólares para la empresa asociado a los puntos redimidos.

##  🌍Customer Loyalty History.csv


- **Loyalty Number**  
  Identificador único del cliente (clave para unir ambos datasets).

- **Country**
  País de residencia del cliente.

- **Province**
  Provincia o estado de residencia del cliente

- **City**
  Ciudad de residencia del cliente.

  - **Postal Code**  
  Código postal.

- **Gender**  
  Género del cliente (Male: masculino y Female: femenino).

- **Education**  
  Nivel educativo alcanzado.

- **Salary**  
  Ingreso anual estimado del cliente.

- **Marital Status**  
  Estado civil (Single,Married,Divorced etc)

- **Loyalty Card**  
  Tipo o nivel de tarjeta dentro del programa (puede reflejar categorías como Silver, Gold, etc.).

- **CLV (Customer Lifetime Value)**  
  Valor total estimado que el cliente genera para la empresa durante toda su relación.

- **Enrollment Type**  
  Tipo de inscripción en el programa (ej. Standard).

- **Enrollment Year**
  Año en el que el cliente se unió al programa.

- **Enrollment Month**  
  Mes en el que el cliente se unió al programa.

- **Cancellation Year** 
  Año en que el cliente canceló su membresía

- **Cancellation Month**  
  Mes en que el cliente canceló su membresía
