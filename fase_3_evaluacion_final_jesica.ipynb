```python
#1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?
df_final.groupby('month')['flights_booked'].sum().reset_index().sort_values('flights_booked',ascending=False)
```

```python
#Lo vemos en un gráfico:Un barplot es una buena opción para analizar variables numéricas y categóricas
sns.barplot(data=df_final, x='month', y='flights_booked',errorbar=None)

plt.title('Vuelos reservados por mes')
plt.xlabel('Mes')
plt.ylabel('Número de vuelos')

plt.show()
```

En este gráfico se ve que el mes con más vuelos reservados es el mes de Julio seguido de agosto y de junio.Puede deberse a las vacaciones de verano. Por otro lado,
En el mes de dicciembre se ve que hay un pico de vuelos reservados, esto puede deberse a que es el mes de navidad.

```python
#2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente?
#Utilizamos un scatterplot para analizar las variables distancia y puntos acumulados ya que es uno se los recomenados para ver la relación entre dos variables numéricas.
# la relación entre dos variables numéricas
sns.scatterplot(data=df_final, x='distance', y='points_accumulated');
```

```python
#En el scatterplot vemos que hay una relación positiva entre los puntos acumulados y la distancia de los vuelos.
# Por lo que cuanto más largo es el vuelo(la distancia),más puntos acumula el cliente.
```

```python
#3. ¿Cuál es la distribución de los clientes por provincia o estado?

df_final.groupby('province')['loyalty_number'].sum().reset_index().sort_values('loyalty_number',ascending=False)
```

```python
#Para visualizarlo, vamos a utilizar un barplot, ya que es una buena opción para analizar una variable numérica y una variable categórica.
#Cada barra representa el promedio de clientes por provincia.
sns.countplot(data=df_final, y='province',palette='Set2')
plt.title('Distribución clientes por provincia')
plt.xlabel('Numero de clientes')
plt.ylabel('Provincia')
plt.xticks(rotation=45); # para que las etiquetas del eje x se vean mejor

```

```python
#Vemos que la provincia con más clientes es Ontario, seguida  British Columbia y Quebec.
# Esto puede deberse a que estas provincias tienen una mayor población o a que hay más vuelos disponibles desde estas provincias.
```

```python
#4. ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?
df_final.groupby('education')['salary'].mean().reset_index().sort_values('salary',ascending=False).round(2)
```

```python
sns.barplot(data=df_final,x='education',y='salary',palette='Set2' ,errorbar=None)

plt.title('Salario promedio por nivel educativo')
plt.xlabel('Nivel educativo')
plt.ylabel('Salario promedio')
plt.xticks(rotation=45)  # girar etiquetas para mejor visibilidad
plt.show()
```

```python
#Como podemos ver los que más ganan son los clientes con un doctorado, seguidos de los que tienen un máster y por último los que tienen un Bachellor
```

```python
#5. ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?
sns.barplot(data=df_final, y='loyalty_number', x='loyalty_card', palette='Set2',errorbar=None)
plt.title('Proporción de clientes por tipo de tarjeta de fidelidad')
plt.xlabel('Tipo de tarjeta')
plt.ylabel('Proporción de clientes')




```

```python
#6. ¿Cómo se distribuyen los clientes según su estado civil y género?

sns.countplot(data=df_final, x='marital_status', hue='gender', palette='Set2')
plt.title('Distribución de clientes por estado civil y género')
plt.xlabel('Estado civil')
plt.ylabel('Cantidad de clientes')
plt.xticks(rotation=45)
plt.show()
```
