```python
#1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?
df_final.groupby('month')['flights_booked'].sum().reset_index().sort_values('flights_booked',ascending=False)
```

```python
#Lo vemos en un gráfico:Un barplot es una buena opción para analizar variables numéricas y categóricas
sns.barplot(data=df_final, x='month', y='flights_booked',errorbar=None)

plt.title('Vuelos reservados por mes')
plt.xlabel('Mes')
plt.ylabel('Número de vuelos')

plt.show()
```

En este gráfico se ve que el mes con más vuelos reservados es el mes de Julio seguido de agosto y de junio.Puede deberse a las vacaciones de verano. Por otro lado,
En el mes de dicciembre se ve que hay un pico de vuelos reservados, esto puede deberse a que es el mes de navidad.

```python
#2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente?
#Utilizamos un scatterplot para analizar las variables distancia y puntos acumulados ya que es uno se los recomenados para ver la relación entre dos variables numéricas.
# la relación entre dos variables numéricas
sns.scatterplot(data=df_final, x='distance', y='points_accumulated');
```

```python
#En el scatterplot vemos que hay una relación positiva entre los puntos acumulados y la distancia de los vuelos.
# Por lo que cuanto más largo es el vuelo(la distancia),más puntos acumula el cliente.
```

```python
#3. ¿Cuál es la distribución de los clientes por provincia o estado?

df_final.groupby('province')['loyalty_number'].sum().reset_index().sort_values('loyalty_number',ascending=False)
```

```python
df_final.groupby('province')['loyalty_number'].sum().reset_index().sort_values('loyalty_number',ascending=False)
```

```python
#Para visualizarlo, vamos a utilizar un barplot, ya que es una buena opción para analizar una variable numérica y una variable categórica.
sns.barplot(data=df_final, x='province', y='loyalty_number',errorbar=None,palette='viridis')
plt.title('Número de clientes por provincia')
plt.xlabel('Provincia')
plt.ylabel('promedio de clientes')
plt.xticks(rotation=45); # para que las etiquetas del eje x y que se vean mejor
```
